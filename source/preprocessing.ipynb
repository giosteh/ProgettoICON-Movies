```python
import pandas as pd
```

### Preprocessing di "movies.csv"

```python
df = pd.read_csv('../dataset/movies.csv')

df.info()
```

```python
# rinomina la colonna 'name' in 'title'
df = df.rename(columns={'name': 'title'})

# rimuovi la colonna 'released'
df = df.drop('released', axis=1)

# rimuovi le righe con budget e gross nulli
df = df[(df['budget'].notnull()) & (df['gross'].notnull())]

df.info()
```

```python
# aggiungi in testa una colonna 'id' con un indice progressivo per ogni film
df.insert(0, 'id', range(1, 1 + len(df)))
```

```python
# stampa le righe senza rating
df[df['rating'].isnull()]
```

```python
# riempi i valori nulli della colonna 'rating'
df.loc[df['title'] == 'Nausicaä of the Valley of the Wind', 'rating'] = 'PG'
df.loc[df['title'] == 'Y Tu Mamá También', 'rating'] = 'R'
df.loc[df['title'] == 'Return of the Living Dead III', 'rating'] = 'R'
df.loc[df['title'] == 'Jason Goes to Hell: the Final Friday', 'rating'] = 'R'
df.loc[df['title'] == 'Madadayo', 'rating'] = 'PG'
df.loc[df['title'] == 'Happy Together', 'rating'] = 'R'
df.loc[df['title'] == 'Eyes Wide Shut', 'rating'] = 'R'
df.loc[df['title'] == 'Brother', 'rating'] = 'R'
df.loc[df['title'] == 'Monster\'s Ball', 'rating'] = 'R'
df.loc[df['title'] == 'Ichi the Killer', 'rating'] = 'R'
df.loc[df['title'] == '31', 'rating'] = 'R'
df.loc[df['title'] == 'Clinton Road', 'rating'] = 'R'
```

```python
# riempi i valori nulli della colonna 'country'
df.loc[df['title'] == 'Clinton Road', 'country'] = 'United States'

# riempi i valori nulli della colonna 'company'
df.loc[df['title'] == 'The Bear', 'company'] = 'Emabassy Pictures'
df.loc[df['title'] == 'Heart of Dixie', 'company'] = 'Orion Pictures'

# riempi i valori nulli della colonna 'runtime'
df.loc[df['title'] == 'One for the Money', 'runtime'] = 91

df.info()
```

```python
# rimuovi i duplicati
df.drop_duplicates(subset=['title', 'director', 'year'], keep='first', inplace=True)

df.info()
```

```python
df.to_csv('../dataset/movies_v2.csv', index=False)
```

### Preprocessing di "combined.csv"

```python
df = pd.read_csv('../dataset/combined.csv')

df.info()
```

```python
# rinomina la colonna 'primaryName' in 'name'
df = df.rename(columns={'primaryName': 'name'})
```

```python
# trasforma la colonna 'primaryProfession' in 3 colonne separate
df[['profession1', 'profession2', 'profession3']] = df['primaryProfession'].str.split(',', expand=True)

df['profession1'] = df['profession1'].str.strip()
df['profession2'] = df['profession2'].str.strip()
df['profession3'] = df['profession3'].str.strip()

# rimuovi la colonna 'primaryProfession'
df = df.drop('primaryProfession', axis=1)

df.info()
```

```python
# riempi i valori nulli di 'profession2' e 'profession3' con 'none'
df['profession2'] = df['profession2'].fillna('none')
df['profession3'] = df['profession3'].fillna('none')

df.info()
```

```python
# sostituisci il valore 'actress' con 'actor' nelle colonne 'profession1', 'profession2' e 'profession3'
df.loc[df['profession1'] == 'actress', 'profession1'] = 'actor'
df.loc[df['profession2'] == 'actress', 'profession2'] = 'actor'
df.loc[df['profession3'] == 'actress', 'profession3'] = 'actor'
```

```python
# riempi i valori nulli di 'deathYear' con 0 e cambia il tipo di 'deathYear' in int64
df['deathYear'] = df['deathYear'].fillna(0)
df['deathYear'] = df['deathYear'].astype('int64')

df.info()
```

```python
df.to_csv('../dataset/combined_v2.csv', index=False)
```
