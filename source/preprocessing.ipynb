```python
df = pd.read_csv('../dataset/combined.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df[['profession1', 'profession2', 'profession3']] = df['primaryProfession'].str.split(',', expand=True)

df['profession1'] = df['profession1'].str.strip()
df['profession2'] = df['profession2'].str.strip()
df['profession3'] = df['profession3'].str.strip()

df = df.drop('primaryProfession', axis=1)
```

```python
df.info()
```

```python
# sostituisci i valori nulli delle colonne profession2 e profession3 con la stringa "none"
df['profession2'] = df['profession2'].fillna('none')
df['profession3'] = df['profession3'].fillna('none')

df.info()
```

```python
# sostituisci i valori nulli della colonna deathYear con 0
df['deathYear'] = df['deathYear'].fillna(0)
```

```python
# cambia il formato della colonna "deathYear" con il formato int64
df['deathYear'] = df['deathYear'].astype('int64')

df.info()

df.head()
```

```python
# salva il dataframe in un nuovo file chiamato: "combined_v2"
df.to_csv('../dataset/combined_v2.csv', index=False)
```

```python
# carica i dataset movies_v2 e combined_v2
movies = pd.read_csv('../dataset/movies_v2.csv')
combined = pd.read_csv('../dataset/combined_v2.csv')
```

```python
# cancella tutte le righe di combined che hanno un valore di deathYear < 1980
combined = combined[(combined['deathYear'] >= 1980) | (combined['deathYear'] == 0)]
```

```python
combined = combined.drop_duplicates(subset=['primaryName', 'birthYear'], keep='first')
combined.info()

movies = movies.drop_duplicates(subset=['name', 'year', 'director'], keep='first')
movies.info()
```

```python
df_join = pd.merge(movies, combined, left_on='star', right_on='primaryName', how='inner')
```

```python
# salva il dataframe in un nuovo file chiamato: "movies_combined"
df_join.to_csv('../dataset/movies_combined.csv', index=False)
```
